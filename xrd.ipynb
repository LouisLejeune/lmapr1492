```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator 
import matplotlib.pyplot as plt 
m = MPRester("JsnJ9pas7aGyFxGDx")
#Structure for material id
structure =m.get_structure_by_material_id("mp-23009")

#Utilisation de XRDCalculator
xdrdif = XRDCalculator(wavelength="CuKa")
pat = xdrdif.get_pattern(structure)

print("Premier pic de dif du diffractogramme :")
print('\n')
print("Angle 2\u03B8 :", pat.x[0])
print("Intensité du pic :", pat.y[0])
print("Indice hkl:", pat.hkls[0])
print('\n')

print("Deuxième pic de dif du diffractogramme :")
print('\n')
print("Angle 2\u03B8 :", pat.x[1])
print("Intensité du pic :", pat.y[1])
print("Indice hkl:", pat.hkls[1])
print('\n')

print("Troisieme pic de dif du diffractogramme :")
print('\n')
print("Angle 2\u03B8 :", pat.x[2])
print("Intensité du pic :", pat.y[2])
print("Indice hkl:", pat.hkls[2])
print('\n')

print("Graphe des trois premiers pics du diffractogramme en fonction de 2\u03B8:")
xdrdif.show_plot(structure,annotate_peaks=True,two_theta_range=[0,60], with_labels=True)
```
