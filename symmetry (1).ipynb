```python
import pymatgen as mg
from pymatgen.io.cif import CifWriter
from pymatgen.ext.matproj import MPRester
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
import matplotlib.pyplot as plt
from numpy import *
from pymatgen.core.operations import *
import pymatgen.symmetry.analyzer as sa
m = MPRester("JsnJ9pas7aGyFxGDx")

# Structure for material id
structure = m.get_structure_by_material_id("mp-23009")

#Trouver les symétries
s = SpacegroupAnalyzer(structure)
Sym = s.get_symmetry_operations()

#Coordonnées des différents atomes de Ca2BrN grace à materialsproject
Br =[0,0,0]
Ca0=[0.2254,0.2254,0.2254]
Ca1=[0.7746,0.7746,0.7746]
N = [0.5,0.5,0.5]

#Première opération de symétrie
Sym1 = SymmOp.from_xyz_string('-x,-y,-z')
Br1 = Sym1.operate(Br)
Ca2 = Sym1.operate(Ca0)
Ca3 = Sym1.operate(Ca1)
N1 = Sym1.operate(N)
print("atome d entree :", Ca0)
print("atome d arrivée :", Ca2)
print('operation de symetrie : \n', Ca2, '=', Sym1.translation_vector, "+ \n", Sym1.rotation_matrix, ".", Ca0)
print()

Image1 = img.imread('Devoir2.PNG')
fig,axes = plt.subplots(figsize=(9,9))
axes.imshow(Image1)
plt.show

#Deuxième opération de symétrie
Sym2  = SymmOp.from_xyz_string('-y-1/3,-x+1/3,z+1/3')
Br1 = Sym2.operate(Br)
Ca2 = Sym2.operate(Ca0)
Ca3 = Sym2.operate(Ca1)
N1 = Sym2.operate(N)
print("atome d entree :", Br)
print("atome d arrivée :", Br1)
print('operation de symetrie : \n', Br1, '=', Sym2.translation_vector, "+ \n", Sym2.rotation_matrix, ".", Br)
print()

Image2 = img.imread('Devoir1.PNG')
fig,axes = plt.subplots(figsize=(9,9))
axes.imshow(Image2)
plt.show

#Troisième opération de symétrie
Sym3  = SymmOp.from_xyz_string('-y,-x,z')
Br1 = Sym3.operate(Br)
Ca2 = Sym3.operate(Ca0)
Ca3 = Sym3.operate(Ca1)
N1 = Sym3.operate(N)
print("atome d entree :", N1)
print("atome d arrivée :", N)
print('operation de symetrie : \n', N1, '=', Sym3.translation_vector, "+ \n", Sym3.rotation_matrix, ".", N)
print()

Image2 = img.imread('Devoir01.PNG')
fig,axes = plt.subplots(figsize=(9,9))
axes.imshow(Image2)
plt.show
```
